# Drop a specific column

```python
df = df.drop(columns = ['Total'])
df.head()
```

```python
# Another way to get the totals
#Here we used for all rows so (:) and for the columns we used (4(inclusive) and 10(exclusive means till col 9))
# axis = 1 means we are adding horizontaly and axis = 0 means vertically
df['Total'] = df.iloc[:,4:10].sum(axis = 1)
df.head()
```

# Filtering Data

```python
df.loc[(df['Type 1']=='Grass')]
```

```python
df.loc[(df['Type 1']=='Grass') & (df['Type 2'] == 'Poison')] #Here I have Used AND condition
```

```python
df.loc[(df['Type 1']=='Grass') | (df['Type 2'] == 'Poison')] #Here I have Used OR condition
```

```python
#To store the filetered data in New DataFrame
new_df = df.loc[(df['Type 1']=='Grass') & (df['Type 2'] == 'Poison') & (df['HP'] > 70)]
new_df
```

```python
#So in the above data we can see that the indexation is wrong to make it right we have to use 
```

```python
#To reset index
new_df = new_df.reset_index()
new_df
```

```python
#As in the above we can see that the old index has been saved as a new column so to drop that
new_df = new_df.reset_index(drop = True)
new_df
```

```python
#Instead of every time assigning it to a DF we can simply use inplace = True to give correct index 
new_df.reset_index(drop = True, inplace=True)
new_df
```

```python
#To display all the names that Contains Mega
df.loc[df['Name'].str.contains('Mega')]
```

```python
#To display all the names that doesnot Contains Mega
df.loc[~df['Name'].str.contains('Mega')]
```

```python
# Suppose we want to filter data using contains where Type 1 = Grass or Fire 
#For this we will use regex 
import re 
df.loc[df['Type 1'].str.contains('Fire|Grass', regex= True)]
```

```python
# The above one is Case Sensitive to ignore this we will use flags
df.loc[df['Type 1'].str.contains('fire|grass',flags = re.I, regex= True)]
```

```python
# I want the Pokemon Names starts with Pi
# ^ means the name starts with pi
# [a-z] means it can have anything within A to Z
# * means it ends with 0 or more characters
df.loc[df['Name'].str.contains('^pi[a-z]*',flags = re.I, regex= True)]
```

# Conditional Changes

```python
# Want to change Fire to Flamer in Type 1 Column
df.loc[df['Type 1'] == 'Fire', 'Type 1'] = 'Flamer'
df
```

```python
# Want to change Fire Pokemons Legendary = True
df.loc[df['Type 1'] == 'Fire', 'Legendary'] = True
df
```

```python
#If you want to change multiple values
df.loc[df['Total']>400, ['Generation','Legendary']] = 'Test Value'
df
```

```python
#If you want to change multiple values for multiple columns(unique)
df.loc[df['Total']>400, ['Generation','Legendary']] = '34', 'Test 2'
df
```

# Working with Large set of Data

```python
for df in pd.read_csv('modified_data.csv', chunksize=50):
    print("Chunk DF")
    print(df)
```

```python
new_df = pd.DataFrame(columns=df.columns)
new_df
```

```python
for df in pd.read_csv('modified_data.csv',chunksize=5):
    results = df.groupby(['Type 1']).count()
    
    new_df = pd.concat([new_df,results])
    
```
